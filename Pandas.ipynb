```python
import pandas as pd

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, 'np.nan', 9, 20, 14.5, 'np.nan', 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes'],
'labels' : ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']}

exam_data_df = pd.DataFrame(exam_data) #to obtain dictionary on dataframe datasructure
exam_data_index = exam_data_df.set_index('labels') #to set index to labels
print(exam_data_index)

exam_data_getrows = exam_data_index.head() #to obtain 5 rows
exam_data_rows = exam_data_getrows[0:3] #to obtain 3 rows
print(exam_data_rows)


exam_data_nonnanscores = exam_data_index.drop(['d','h'], axis = 0) #to drop rows with np.nan score
print(exam_data_nonnanscores)

exam_data_extract = exam_data_nonnanscores.iloc[:,0:2] #to extract new columns
print(exam_data_extract)

exam_data2 = {'name': ['Suresh'], 'score': [15.5], 'attempts': [1], 'qualify': ['yes']} #to add new data
exam_data_index2 = pd.DataFrame(exam_data2)
new_exam_data_index = pd.concat([exam_data_nonnanscores, exam_data_index2], axis = 0) #to concatenate 2 dataframes
#new_exam_data_index = new_exam_data_index.set_index('labels') #to set index to labels
print(new_exam_data_index)

exam_data_remove = new_exam_data_index.drop('attempts', axis ='columns')
print(exam_data_remove)

z = new_exam_data_index['score']
Success = list() #to create a list for Success column

for element in z:
    if element > 10:
        Success.append(1) #to append 1 or 0 into Success
    else:
        Success.append(0) 

exam_data_index3 = pd.Series(Success) #to create column
new_exam_data_index3 = pd.DataFrame(exam_data_index3)
new_exam_data_index3 = new_exam_data_index3.rename(columns = {0: 'Success'})
new_exam_data_index4 = pd.concat([new_exam_data_index, new_exam_data_index3], axis = 1) #to concatenate Success column to existing dataframe datastructure
print(new_exam_data_index4)


new_exam_data_index4.to_csv("my_data.csv")
new_data = pd.read_csv("my_data.csv")
```
